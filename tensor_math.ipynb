```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
tensor_a + tensor_b
```

```python
#Addition long hand
torch.add(tensor_b,tensor_a)
```

```python
tensor_b - tensor_a
```

```python
#sub function
torch.sub(tensor_b,tensor_a)
```

```python
tensor_a * tensor_b
```

```python
#MUl longhand
torch.mul(tensor_b,tensor_a)
```

```python
#Division
tensor_b / tensor_a  #torch.div()
```

```python
tensor_b % tensor_a  #torch.reminder
```

```python
torch.pow(tensor_a,tensor_b)
```

```python
tensor_a.add(tensor_b)
```

```python
#re assignemt
tensor_a = tensor_a+tensor_b
#or
tensor_a.add_(tensor_b)
```
